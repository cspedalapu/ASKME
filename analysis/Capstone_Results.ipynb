```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load your spreadsheet
df = pd.read_excel("Chandrasekhar_Human_Evaluation_Questions_GPT4o.xlsx")

# Step 2: Calculate the mean of each score column
mean_scores = df[['Fluency', 'Coherence', 'Relevance', 'Factuality']].mean()

# Step 3: Plot the bar chart
plt.figure(figsize=(8, 6))
mean_scores.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Average Scores by Category')
plt.ylabel('Mean Score')
plt.ylim(0, 5)  # Since scores range from 1 to 5
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# Calculate mean values of the four scoring columns
mean_values = df[['Fluency', 'Coherence', 'Relevance', 'Factuality']].mean()

print(mean_values)
```

```python
# Group by question and calculate the mean scores
grouped_means = df.groupby('Question Number')[['Fluency', 'Coherence', 'Relevance', 'Factuality']].mean()

# Sort by Factuality
sorted_by_factuality = grouped_means.sort_values(by='Factuality', ascending=False)

# Get top 10 and bottom 10
top_10 = sorted_by_factuality.head(10)
bottom_10 = sorted_by_factuality.tail(10)

# Plot top 10
plt.figure(figsize=(12, 6))
top_10.plot(kind='bar', colormap='magma', ax=plt.gca())
plt.title('Top 10 Questions by Factuality')
plt.xlabel('Question Number')
plt.ylabel('Mean Score')
plt.ylim(0, 5)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Plot bottom 10
plt.figure(figsize=(12, 6))
bottom_10.plot(kind='bar', colormap='magma', ax=plt.gca())
plt.title('Bottom 10 Questions by Factuality')
plt.xlabel('Question Number')
plt.ylabel('Mean Score')
plt.ylim(0, 5)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Filter for Question Number 1
q1_df = df[df['Question Number'] == 1]

# Calculate mean scores for Question 1
q1_mean = q1_df[['Fluency', 'Coherence', 'Relevance', 'Factuality']].mean()

# Plot as bar chart
plt.figure(figsize=(6, 4))
q1_mean.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Average Scores for Question 1')
plt.ylabel('Mean Score')
plt.ylim(0, 5)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Group by Question Number and compute mean scores
question_means = df.groupby('Question Number')[['Fluency', 'Coherence', 'Relevance', 'Factuality']].mean()

# Plot boxplots of average scores across all questions
plt.figure(figsize=(8, 6))
question_means.boxplot()

plt.title('Distribution of Average Scores per Question')
plt.ylabel('Average Score')
plt.ylim(0.5, 5.5)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
